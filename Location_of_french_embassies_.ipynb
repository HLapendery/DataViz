# **Plotting french embassies location from [french government data](https://www.data.gouv.fr/fr/datasets/geolocalisation-des-ambassades-de-france/)**

# Importing libraries

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import pandas as pd
import os
```

# Formating data

```python
path = os.getcwd()
path = "D:\hippo\Documents\PYTHON\DATA\gouv"
file = "geolocalisation-ambassades-de-france.csv"
csv_file = path+"\\"+file
data = pd.read_csv(csv_file, sep=";")
data = data.iloc[:,1:]
data.head()
```

# Creating map and plotting data

```python
m=Basemap(llcrnrlon=-170, llcrnrlat=-60,urcrnrlon=190,urcrnrlat=90, lon_0=0)
plt.figure(figsize=(12,8))

m.drawmapboundary(fill_color='#A6CAE0', linewidth=0, zorder=0)
m.fillcontinents(color='grey', alpha=.8, lake_color='grey', zorder=1)
m.drawcoastlines(linewidth=0.1, color="white")
m.drawcountries(linewidth=0.2, color="white")

parislat, parislon = 48.866667,2.333333
           
m.scatter(data["longitude"], data["latitude"], c='r', zorder=2,label = 'Embassies ')
m.scatter(parislon,parislat,c = 'g', zorder=2, label = 'Paris')

plt.legend(loc='lower left')
plt.title("Location of french embassies around the world",fontsize=24)
plt.savefig("images/rembassies_loc.png")
plt.show()
```
