# Every Mad Men epidodes Imdb rating

# Importing libraries

```python
from bs4 import BeautifulSoup as bs
import requests as rq
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

## Scraping data

```python
url = "https://www.imdb.com/title/tt0804503/episodes?season="
all_ratings = []
for i in tqdm(range(1,8)):
    season_ratings = []
    url_season = url + str(i)
    page_season = rq.get(url_season)
    html = bs(page_season.text,"html.parser")
    divs = html.find_all("div",class_ = "ipl-rating-star small")
    for div in divs:
        rating = [r.text for r in div.find_all("span",class_ = "ipl-rating-star__rating")]
        season_ratings.append(float(rating[0]))
    all_ratings.append(season_ratings)
df_seasons = pd.DataFrame(all_ratings)
df_seasons.columns = range(1,15)
df_seasons.index = range(1,8)
```

# Plotting data

```python
df_seasons
plt.figure(figsize=(14,7))
sns.heatmap(df_seasons,annot=True, vmax=10, linewidths=.2,cmap = 'coolwarm', cbar = False)
```
