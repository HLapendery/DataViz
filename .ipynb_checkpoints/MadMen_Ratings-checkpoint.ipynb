# Every Mad Men epidodes Imdb rating

## Importing libraries

```python
%matplotlib inline

from bs4 import BeautifulSoup as bs
import requests as rq
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

plt.style.use('ggplot')
```

## Scraping data

```python
url = "https://www.imdb.com/title/tt0804503/episodes?season="
all_ratings = []
for i in tqdm(range(1,8)):
    season_ratings = []
    url_season = url + str(i)
    page_season = rq.get(url_season)
    html = bs(page_season.text,"html.parser")
    divs = html.find_all("div",class_ = "ipl-rating-star small")
    for div in divs:
        rating = [r.text for r in div.find_all("span",class_ = "ipl-rating-star__rating")]
        season_ratings.append(float(rating[0]))
    all_ratings.append(season_ratings)
df_seasons = pd.DataFrame(all_ratings)
df_seasons.columns = range(1,15)
df_seasons.index = range(1,8)
```

## Plotting data

```python
plt.figure(figsize=(14,7))
axs = plt.gca()
sns.heatmap(df_seasons,annot=True, vmax=10, square = True, linewidths=.1,cmap = 'coolwarm', cbar = False)
plt.xlabel("Episodes",labelpad=10, color = 'grey', fontsize= 14)
plt.ylabel("Seasons",rotation=0,labelpad=40, color = 'grey', fontsize= 14)
axs.xaxis.set_ticks_position('top')
axs.xaxis.set_label_position('top')
axs.yaxis.set_tick_params(length = 0)
axs.xaxis.set_tick_params(length = 0)
plt.xticks(color='grey', size=14)
plt.yticks(color='grey', size=14)
plt.yticks(rotation=0)
```

```python
df_seasons.head()
fig,axs = plt.subplots(1,2,figsize=(15,5),sharey=True)  
fig.suptitle("Mad Men Imdb Rating",fontsize=20, y=1.1)
sns.boxplot(data=df_seasons, ax=axs[0])
sns.boxplot(data=df_seasons.T, ax=axs[1])

axs[0].set_title("by episodes")
axs[1].set_title("by seasons")
plt.tight_layout()
```
